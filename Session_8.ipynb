```r
 library(readxl)
```

```r
da=read_excel("macro.xls")
```

```r
da=data.frame(da)
```

```r
View(da)
```

```r
summary(da)
```

```r
msoft=da[,2]
```

```r
sandp=da[,3]
```

```r
cpi=da[,4]
```

```r
indpro=da[,5]
```

```r
m1supply=da[,6]
```

```r
ccredit=da[,7]
```

```r
bminusa=da[,8]
```

```r
ustb3m=da[,9]
```

```r
ustb10y=da[,10]
```

```r
rmsoft=c(NA,100*diff(log(msoft))) #computes the log return series for Microsoft
rsandp=c(NA,100*diff(log(sandp))) #computes the log return series for sandp
inflation=c(NA,100*diff(log(cpi))) #computes inflation
dspread=c(NA,100*diff(bminusa)) #computes spread
dcredit=c(NA,100*diff(ccredit)) #computes spread credit
dprod=c(NA,100*diff(indpro)) #computes spread prod
dmoney=c(NA,100*diff(m1supply)) #computes spread money
dinflation=c(NA,100*diff(inflation)) #computes spread inflation
rterm=c(NA,100*diff(ustb10y-ustb3m))
```

```r
ustb3mmonth=ustb3m/12 #transforms the T-bill yields into monthly figures
ersandp=rsandp-ustb3mmonth #generates the excess market return
ermsoft=rmsoft-ustb3mmonth
```

```r
modelapt=lm(ermsoft ~ ersandp + dprod + dcredit + dinflation + dmoney + dspread + rterm, data=da)
```

```r
summary(modelapt)
```

```r
plot(da$Date[-(1:2)], modelapt$residuals,type="l",xlab="Date",ylab="Residuals")
```
